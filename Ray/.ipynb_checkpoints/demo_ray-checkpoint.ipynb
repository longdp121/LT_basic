```python
from io import StringIO
from collections import Counter
from time import sleep
from itertools import combinations
import random
import logging

import ray
import pandas as pd
import requests

print('Done')
```

```python
internet = requests.get('https://moz.com/top-500/download/?table=top500Domains')
```

```python
df = pd.read_csv(StringIO(internet.text))
df.head()
```

```python
ray.init(ignore_reinit_error=True, logging_level=logging.ERROR)
```

```python
@ray.remote
def index(url):
    print(f'Sleep  {url}')
    sleep(.5)
    print(f'Not sleep  {url}')
    return url.split('.')[-1]
```

```python
for url in df['Root Domain']:
    print(index.remote(url))
```

```python
futures = [index.remote(url) for url in df['Root Domain']]
```

```python
ray.get(futures)
```
