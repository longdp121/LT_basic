```python
#Pyplot API
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), label = "sin(x)");
plt.plot(x, np.cos(x), label = "cos(x)");
plt.xlabel("Bien x")
plt.ylabel("Sin(x)")
# plt.xlim([0, 4])
# plt.ylim([-0.75, 1])

plt.axis("equal")
# plt.axis([0, 4, -0.75, 1])

plt.legend();
```

```python
#OO Api
x = np.linspace(0, 10, 100)
y = np.sin(x)
a = np.linspace(0, 10, 50)
b = np.cos(a)
fig, ax = plt.subplots(figsize = (12, 12))
ax.plot(x, y);
ax.set(title = "My plot",
      xlabel = "This x",
      ylabel = "This y");
ax.plot(a, b);
```

```python
#OO Api
#ran = np.random.RandomState(0)
x = [4, 4, 4, 4, 5, 6, 7, 8, 9]
y = [3, 4, 4, 1, 2, 8, 4, 11, 7]
#colors = ran.rand(100)
#sizes = 100 * ran.rand(100)

fg, ax = plt.subplots()
#ax.grid(False)
img = ax.scatter(x, y, cmap = "viridis", alpha = 1);
fg.colorbar(img);
```

```python
#Bar - biu  ct
#Vertical
my_dict_data = {"Cho" : 10,
               "Meo" : 8,
               "Chuot" : 12,
               "Trau" : 17,
               "Lon" : 9}
fig, ax = plt.subplots()
ax.bar(my_dict_data.keys(), my_dict_data.values());
ax.set(title = "Vat nuoi",
      xlabel = "Con gi",
      ylabel = "So");
ax.bar_label(ax.containers[0]);  #Put vallue on top of the bar
```

```python
#Horizotal
fg, ax = plt.subplots()
#Must turn data type into list when using barh
ax.barh(list(my_dict_data.keys()), list(my_dict_data.values()), log = True);
ax.set(title = "Vat nuoi ngang",
      xlabel = "So",
      ylabel = "Con gi");
ax.bar_label(ax.containers[0]);
```

```python
#Histogram
#Th hin tn sut di dang ct
np.random.seed(42)
student_height = np.random.normal(170, 20, 250)
```

```python
#API
plt.hist(student_height, bins = 30);  #Bins for giving space between bar
plt.title("Chieu cao cac ban")
plt.xlabel("Height in mm");
plt.ylabel("Number");
```

```python
#OO API

fig, ax = plt.subplots()
ax.hist(student_height, bins = 30);
ax.set(xlabel = "Height in mm",
      ylabel = "Number",
      title = "Chieu cao cac ban")
ax.bar_label(ax.containers[0]);
```

```python
#Subplots - v nhiu biu  trn cng mt fig
#Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (15, 10))
#Plot data
ax1.plot(x, np.sin(x));
ax1.plot(y, np.cos(y));
a = ax2.scatter(x, y);
ax3.bar((my_dict_data.keys()), (my_dict_data.values()));
ax3.bar_label(ax3.containers[0]);
ax4.hist(student_height);
ax4.bar_label(ax4.containers[0]);
```

```python
#Option 2
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10, 5))
ax[0, 0].plot(x, np.cos(x));
ax[0, 0].plot(y, np.cos(y));

ax[0, 1].scatter(x, y);
ax[1, 0].barh(list(my_dict_data.keys()), list(my_dict_data.values()));
ax[1, 0].bar_label(ax[1, 0].containers[0]);
ax[1, 1].hist(student_height, bins = 40);
ax[1, 1].bar_label(ax[1, 1].containers[0]);
```
