```python
#Pyplot API
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), label = "sin(x)");
plt.plot(x, np.cos(x), label = "cos(x)");
plt.xlabel("Bien x")
plt.ylabel("Sin(x)")
# plt.xlim([0, 4])
# plt.ylim([-0.75, 1])

plt.axis("equal")
# plt.axis([0, 4, -0.75, 1])

plt.legend();
```

```python
#OO Api
x = np.linspace(0, 10, 100)
y = np.sin(x)
a = np.linspace(0, 10, 50)
b = np.cos(a)
fig, ax = plt.subplots(figsize = (12, 12))
ax.plot(x, y);
ax.set(title = "My plot",
      xlabel = "This x",
      ylabel = "This y");
ax.plot(a, b);
```
